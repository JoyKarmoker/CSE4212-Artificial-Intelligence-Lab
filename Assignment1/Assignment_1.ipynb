## Write a Python program to display values of y when 'x' is an independent variable having values [-10, 10, step_size = 0.1] and

## y = wx + b; where 'w' and 'b' are constants

```python
# !pip install SciencePlots
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
x = np.arange(-10, 10.1, 0.1)
w = 6
b = 12
print('x:{}, a:{}, b:{}'.format(x[-5:], w, b))
```

```python
plt.figure(figsize=(17,6))
y = w*x + b
plt.plot(x,y, '-', linewidth=2, color='orange', label='Plot of x, y')
plt.plot(x,y, '.', color='purple', ms=4, label='Value of x, y')
plt.grid(color='blue', linestyle='--', linewidth=0.5)
plt.xlabel('-->x')
plt.ylabel('-->y')
plt.title('plot of: wx+b')
plt.legend(loc='lower right')
plt.ylim(top=80)
for i in range(0, len(x), 20):
  plt.annotate(f'{x[i]:.1f}, {y[i]:.1f}', (x[i], y[i]), ha='center', textcoords = 'offset points', xytext=(0,10))
plt.show()
```

## y = x^2

```python
plt.figure(figsize=(17,6))
y = np.power(x, 2)
plt.plot(x,y, '-', linewidth=2, color='orange', label='Plot of x, y')
plt.plot(x,y, '.', color='purple', ms=4, label='Value of x, y')
plt.grid(color='blue', linestyle='--', linewidth=0.5)
plt.xlabel('-->x')
plt.ylabel('-->y')
plt.title('plot of: y = x^2')
plt.legend(loc='lower right')
plt.ylim(top=115)
for i in range(0, len(x), 20):
  plt.annotate(f'{x[i]:.1f}, {y[i]:.1f}', (x[i], y[i]), ha='center', textcoords = 'offset points', xytext=(0,10))
plt.show()
```

## y = 1/ (1 + e^(-x))

```python
y = 1/(1+np.exp(-x))
plt.figure(figsize=(17,6))
plt.plot(x,y, '-', linewidth=2, color='orange', label='Plot of x, y')
plt.plot(x,y, '.', color='purple', ms=4, label='Value of x, y')
plt.grid(color='blue', linestyle='--', linewidth=0.5)
plt.xlabel('-->x')
plt.ylabel('-->y')
plt.title('plot of: y = 1/ (1 + e^(-x))')
plt.legend(loc='lower right')
plt.ylim(top=1.1)
for i in range(0, len(x), 20):
  plt.annotate(f'{x[i]:.1f}, {y[i]:.1f}', (x[i], y[i]), ha='center', textcoords = 'offset points', xytext=(0,10))
plt.show()
```

## y = (e^x - e^(-x)) / (e^x - e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) - np.exp(-x))
plt.figure(figsize=(17,6))
plt.plot(x,y, '-', linewidth=2, color='orange', label='Plot of x, y')
plt.plot(x,y, '.', color='purple', ms=4, label='Value of x, y')
plt.grid(color='blue', linestyle='--', linewidth=0.5)
plt.xlabel('-->x')
plt.ylabel('-->y')
plt.title('plot of: y = (e^x - e^(-x)) / (e^x - e^(-x))')
plt.legend(loc='lower right')
# plt.ylim(top=1.1)
for i in range(0, len(x), 20):
  plt.annotate(f'{x[i]:.1f}, {y[i]:.1f}', (x[i], y[i]), ha='center', textcoords = 'offset points', xytext=(0,10))
plt.show()
```

## y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
u = w*x + b
y = 1/(1+(np.exp(-u)))
plt.figure(figsize=(17,6))
plt.plot(x,y, '-', linewidth=2, color='orange', label='Plot of x, y')
plt.plot(x,y, '.', color='purple', ms=4, label='Value of x, y')
plt.grid(color='blue', linestyle='--', linewidth=0.5)
plt.xlabel('-->x')
plt.ylabel('-->y')
plt.title('plot of: y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))')
plt.legend(loc='lower right')
plt.ylim(top=1.1)
for i in range(0, len(x), 20):
  plt.annotate(f'{x[i]:.1f}, {y[i]:.1f}', (x[i], y[i]), ha='center', textcoords = 'offset points', xytext=(0,10))
plt.show()
```

## y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u - e^(-u))

```python
u = w*x + b
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) - np.exp(-u))
plt.figure(figsize=(17,6))
plt.plot(x,y, '-', linewidth=2, color='orange', label='Plot of x, y')
plt.plot(x,y, '.', color='purple', ms=4, label='Value of x, y')
plt.grid(color='blue', linestyle='--', linewidth=0.5)
plt.xlabel('-->x')
plt.ylabel('-->y')
plt.title('plot of: y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u - e^(-u))')
plt.legend(loc='lower right')
# plt.ylim(top=1.1)
for i in range(0, len(x), 20):
  plt.annotate(f'{x[i]:.1f}, {y[i]:.1f}', (x[i], y[i]), ha='center', textcoords = 'offset points', xytext=(0,10))
plt.show()
```

## y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))
## y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))
## y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
w1 = 5
b1 = 10
u1 = w1*x + b1
y1 = 1/(1+(np.exp(-u1)))

w2 = 20
b2 = 0.987
u2 = w2*x + b2
y2 = 1/(1+(np.exp(-u2)))

w3 = 200
w4 = 12
b = -0.5
u3 = w3*y1 + w4*y2 + b
y3 = 1/(1+(np.exp(-u3)))


plt.figure(figsize=(17,6))
plt.plot(x,y3, '-', linewidth=2, color='orange', label='Plot of x, y')
plt.plot(x,y3, '.', color='purple', ms=4, label='Value of x, y')
plt.grid(color='blue', linestyle='--', linewidth=0.5)
plt.xlabel('-->x')
plt.ylabel('-->y')
plt.title('plot of: y = g3(f3(v))')
plt.legend(loc='lower right')
plt.ylim(top=1.1)
for i in range(0, len(x), 20):
  plt.annotate(f'{x[i]:.1f}, {y[i]:.1f}', (x[i], y3[i]), ha='center', va ='center' , textcoords = 'offset points', xytext=(0,10))
plt.show()
```
